## Tokenization on Dataset

```python
from transformers import AutoTokenizer
model = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
text = "My Favourite food is masala Dosa"
encoded_text = tokenizer(text)
print(encoded_text)
```

```python
emotion_datasets["train"][:2]
```

```python
def tokenization(batch):
  return tokenizer(batch["text"],padding = True,truncation=True)
```

```python
tokenization(emotion_datasets["train"][:2])
```

```python
type(encoded_text)
```

```python
tokens = tokenizer.convert_ids_to_tokens(encoded_text.input_ids)
tokens
```

```python
tokenizer.convert_tokens_to_string(tokens)
```

## load pretrained model

```python
from transformers import TFAutoModel
model_name = "distilbert-base-uncased"
model = TFAutoModel.from_pretrained(model_name)
```

```python
ex_text = "this is test document"
inputs = tokenizer(ex_text,return_tensors = "tf")
inputs
```

```python
outputs = model(inputs)
outputs
```

```python
token_ids = inputs["input_ids"].numpy()[0]

decode_output = tokenizer.decode(token_ids)
decode_output
```
